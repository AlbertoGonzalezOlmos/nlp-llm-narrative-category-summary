```python
import os
import glob
import openai
import json

def load_api_key(config_file):
    with open(config_file) as f:
        config = json.load(f)
    return config['api_key']

def configure_openai_api(api_key):
    openai.api_key = api_key

def get_text_files(directory):
    return glob.glob(os.path.join(directory, '*.txt'))

def read_file(file_path):
    with open(file_path, 'r') as f:
        content = f.read()
    return content

def generate_summary(text, query, max_length=500):
    response = openai.Completion.create(
        engine="text-davinci-002",
        prompt=f"{query}: {text}",
        max_tokens=max_length,
        n=1,
        stop=None,
        temperature=0.7,
    )
    summary = response.choices[0].text.strip()
    return summary

def summarize_text_files(directory, query, config_file="config.json", output_file=None):
    api_key = load_api_key(config_file)
    configure_openai_api(api_key)

    summaries = {}
    content_master = ""
    counter = 1
    text_files = get_text_files(directory)
    for file_path in text_files:
        file_name = os.path.basename(file_path)
        content = read_file(file_path)
        
        #summary = generate_summary(content, query)
        content_master += f"user {counter}: {content}; \n "
        counter += 1
        
    print(content_master)
    
    summary = generate_summary(content_master, query)
    summaries = summary
    if output_file:
        with open(output_file, 'w') as f:
            json.dump(summaries, f, indent=2)
    else:
        print(json.dumps(summaries, indent=2))

# Usage example
directory = "./textFiles/"
query = "Analyse the opinions of the different users and come up with a pattern in their opinion."
config_file = "config.json"
output_file = "./outputFiles/summaries_4.txt"
summarize_text_files(directory, query, config_file, output_file)



# Used queries
#"What is the common opinion in the statements of these users?"
#Integrate these points, and then write a one-sentence summary.
```
